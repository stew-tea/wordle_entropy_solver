```python
import nltk
nltk.download('words')
nltk.download('wordnet')

from nltk.corpus import words
from wordfreq import word_frequency
import json

# Get all English words
all_words = [w.lower() for w in words.words()]

# Filter and build freq dicts for each length
for length in [5, 6, 7]:
    # Filter to correct length, alpha only, no proper nouns
    filtered = [w for w in all_words 
                if len(w) == length 
                and w.isalpha()
                and w == w.lower()]
    
    # Deduplicate
    filtered = list(set(filtered))
    
    # Get wordfreq scores, drop very rare words
    freq_dict = {}
    for w in filtered:
        f = word_frequency(w, 'en')
        if f > 1e-6:  # minimum threshold — drops truly obscure words
            freq_dict[w] = f
    
    # Sort by frequency descending
    freq_dict = dict(sorted(freq_dict.items(), key=lambda x: x[1], reverse=True))
    
    print(f"en{length}: {len(freq_dict)} words")
    
    with open(f'en{length}_freq.json', 'w') as f:
        json.dump(freq_dict, f)

print("Done!")
```
